```python
import re
import requests
```

```python
def extract_emails_from_url(url):
    
    response = requests.get(url)
    text_content = response.text

    # Regular expression pattern for extracting email addresses
    email_pattern = r'\b[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Za-z]{2,}\b'

    # Finding all email addresses in the text content using the pattern
    emails = re.findall(email_pattern, text_content)

    return emails

url = "https://www.jnu.ac.in/contactus"

# Extracting email addresses from the URL
email_addresses = extract_emails_from_url(url)

# Printing the extracted email addresses
for email in email_addresses:
    print(email)
```

```python
def extract_phone_numbers(url):
    
    response = requests.get(url)

    # Extracting phone numbers using regex pattern
    pattern1 = r"\b(?:\+91)?(?:\d{2}-\d{8}|\d{3}-\d{7}|\d{4}-\d{6})\b"
    pattern2 = r"\b\d{3}[-.]?\d{3}[-.]?\d{4}\b"
    tel_numbers = re.findall(pattern1, response.text)
    mobile_numbers = re.findall(pattern2, response.text)

    # Returning the extracted phone numbers
    return tel_numbers, mobile_numbers

url = "https://www.jnu.ac.in/scss-contact-us"

# Extracting phone numbers from the URL
phone_numbers = extract_phone_numbers(url)

# Printing the extracted phone numbers
for number in phone_numbers:
    print(number)
```

```python
def extract_dates_from_url(url):
    
    response = requests.get(url)
    html_content = response.text
    
    # Defining regular expressions for different date formats
    date_formats = [
        r'\d{4}-\d{2}-\d{2}',         # YYYY-MM-DD
        r'\d{2}-\d{2}-\d{4}',         # DD-MM-YYYY
        r'\d{2}/\d{2}/\d{4}',         # DD/MM/YYYY
        r'\d{2}/\d{2}/\d{2}',         # DD/MM/YY
        r'\d{2}-\d{2}-\d{2}'          # DD-MM-YY
    ]
    
    extracted_dates = []
    
    # Searching for date patterns in the HTML content
    for date_format in date_formats:
        dates = re.findall(date_format, html_content)
        extracted_dates.extend(dates)
    
    return extracted_dates

url = 'https://www.jnu.ac.in/notices'

dates = extract_dates_from_url(url)
for date in dates:
    print(date)
```
